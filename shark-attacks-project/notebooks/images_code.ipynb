```python
import pandas as pd
df = pd.read_csv('clean_shark_data.csv')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=df, x='Sex')
plt.savefig("sex_plot.png", dpi=300, bbox_inches='tight')
```

```python
sns.histplot(df['Age'], bins=20, kde=True)
plt.savefig("age_distribution.png", dpi=300, bbox_inches='tight') 
plt.show()
```

```python
df['Activity'].value_counts().head(10).plot(kind='barh')
plt.savefig("activity_top10.png", dpi=300, bbox_inches='tight')  
plt.show()
```
